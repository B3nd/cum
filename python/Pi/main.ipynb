```python
%matplotlib inline
import matplotlib.pyplot as plt
#plt.style.use('seaborn-whitegrid')
import numpy as np
import random
```

```python
number_total = 0
number_in_circle = 0
def r():
    return random.uniform(0.0, 1.0)
```

```python
fig = plt.figure()
ax = plt.gca()
ax.cla()
ax.set_xlim((-0.5, 0.5))
ax.set_ylim((-0.5, 0.5))
x = np.linspace(-1, 1)
c = plt.Circle((0, 0), 0.5, color='black', fill=False)
ax.add_artist(c)

number = 100000
for i in range(number):
    x = r() - 0.5
    y = r() - 0.5
    radius  = abs(x)**2 + abs(y)**2
    if radius <= 0.25:
        ax.add_artist(plt.Circle((x, y), 0.0001, color='red'))
        number_in_circle += 1
    else:
        ax.add_artist(plt.Circle((x, y), 0.0001, color='blue'))
    number_total +=1
    
plt.plot([], [], ' ', label="Pi is " + str(4 * number_in_circle / number_total))
plt.legend(frameon=True, loc='lower center', ncol=1)
```
